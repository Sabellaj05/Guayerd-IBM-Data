```python
nombre_dupli = {}

for name in dff["Nombre Proveedor"].unique():
    numero_unico = dff[dff["Nombre Proveedor"] == name]["Número Proveedor"].nunique()

    if numero_unico > 1:
        numero_prov = dff[dff["Nombre Proveedor"] == name]["Número Proveedor"].unique()
        nombre_dupli[name] = list(numero_prov)
        
nombre_dupli
```

```python
data_trans = [(name, num) for name, num in nombre_dupli.items()]
```

```python
dfn = pd.DataFrame(data=data_trans, columns=["Nombre", "Numero"])
dfn
```
